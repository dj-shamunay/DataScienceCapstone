Import Python Classes

```python
import pandas as pd
import requests
import numpy as np
import geocoder
import folium
import requests 
import matplotlib.cm as cm
import matplotlib.colors as colors
import json
import xml
from geopy.geocoders import Nominatim 
from bs4 import BeautifulSoup
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Pittsburgh_neighborhoods"
extracting_data = requests.get(url).text
wiki_data = BeautifulSoup(extracting_data, 'lxml')
```

```python
column_names = ['Suburb']
pittsburgh = pd.DataFrame(columns = column_names)
content = wiki_data.find('div', class_='mw-parser-output')
table = content.table.tbody
```

```python
df = pittsburgh.groupby('Suburb').apply(', '.join).reset_index()
df.head()
```

```python
def neighborhood_list(grouped):    
    return ', '.join(sorted(grouped['Suburb'].tolist()))
                    
grp = df.groupby(['Suburb'])
df_2 = grp.apply(neighborhood_list).reset_index(name='Suburb')
```

```python

address = 'Pittsburgh,Pennsylvania'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude_x = location.latitude
longitude_y = location.longitude


map_pitt = folium.Map(location=[latitude_x, longitude_y], zoom_start=10)

for lat, lng, nei in zip(df['Suburb']):
    
    label = '{}'.format(nei)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_pitt)  
    
map_pitt
```
