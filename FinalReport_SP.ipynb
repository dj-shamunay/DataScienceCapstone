Final Report- The Battle of the Neighborhoods- Pittsburgh, PA

The purpose of this Project is to help people in exploring better facilities around their neighborhood. It will help people making smart and efficient decision on selecting great neighborhood out of numbers of other neighborhoods in Pittsburgh.

This project is for those people who are looking for better neighborhoods. For ease of accessing to resteraunts, School, grocery stores, mall, theatre, hospitals.

This Project aim to create an analysis of features for a people migrating to Pittsburgh to search a best neighborhood as a comparative analysis between neighborhoods. The features include median housing price and better school according to ratings, crime rates of that particular area.
