The Battle of the Neighborhoods Introduction

    The purpose of this project is to compare the Neighborhoods found in Pittsburgh, PA. It will be helpful in helping people select a place to reside when moving to the Pittsburgh Area.

    many categories of data will be used, including population, crime, schools, and resteraunts.
   

    Pittsburgh is a very popular city with many suburbs and neighborhoods. 
   

Libraries used: 
    
    Pandas: for creating dataframes.
    geocoder- to retrieve location data
    matplotlib- for plotting
